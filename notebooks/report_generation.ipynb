```python
# ============================================================
# ADAPTIVE ML-BASED LOG ANOMALY DETECTION SYSTEM
# MASTER PDF REPORT GENERATION PIPELINE
# ============================================================

# RUN ONLY ONCE
# !pip install reportlab pillow

# ==============================
# IMPORTS
# ==============================

import os
import pandas as pd

from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Image
)

from reportlab.lib.styles import (
    getSampleStyleSheet
)

from reportlab.lib.pagesizes import letter

from reportlab.platypus.flowables import (
    HRFlowable
)

from reportlab.lib.units import inch

# ============================================================
# PATH CONFIGURATION
# ============================================================

BASE_DIR = "E:/LogAnomalyDetector/notebooks"

GRAPH_DIR = (
    f"{BASE_DIR}/graphs/advanced_evaluation"
)

REPORT_DIR = (
    f"{BASE_DIR}/reports"
)

os.makedirs(
    REPORT_DIR,
    exist_ok=True
)

PDF_PATH = (
    f"{REPORT_DIR}/adaptive_ml_soc_report.pdf"
)

# ============================================================
# PDF DOCUMENT
# ============================================================

doc = SimpleDocTemplate(

    PDF_PATH,

    pagesize=letter,

    rightMargin=40,
    leftMargin=40,

    topMargin=40,
    bottomMargin=30
)

styles = getSampleStyleSheet()

title_style = styles["Title"]

heading_style = styles["Heading1"]

body_style = styles["BodyText"]

elements = []

print("✅ Report Generator Initialized")

# ============================================================
# TITLE PAGE
# ============================================================

title = Paragraph(

    """
    <font size=22>
    <b>Adaptive ML-Based Log Anomaly Detection System</b>
    </font>
    """,

    title_style
)

subtitle = Paragraph(

    """
    <font size=14>
    Multi-Dataset SOC-Oriented Machine Learning Evaluation Report
    </font>
    """,

    body_style
)

overview = Paragraph(

    """
    This report presents the design, implementation,
    evaluation, and analysis of an Adaptive Machine
    Learning-Based Log Anomaly Detection System
    developed for Security Operations Center (SOC)
    environments.

    The system integrates semantic TF-IDF feature
    extraction, structured behavioral analysis,
    probability-based anomaly scoring, threshold
    optimization, cross-dataset evaluation,
    calibration analysis, and advanced visualization
    techniques for cybersecurity anomaly detection.
    """,

    body_style
)

elements.append(title)

elements.append(
    Spacer(1, 0.3 * inch)
)

elements.append(subtitle)

elements.append(
    Spacer(1, 0.4 * inch)
)

elements.append(
    HRFlowable(width="100%")
)

elements.append(
    Spacer(1, 0.3 * inch)
)

elements.append(overview)

elements.append(
    Spacer(1, 0.5 * inch)
)

# ============================================================
# DATASET SECTION
# ============================================================

dataset_heading = Paragraph(
    "1. Dataset Analysis",
    heading_style
)

dataset_text = Paragraph(

    """
    The anomaly detection pipeline was evaluated
    using multiple publicly available log datasets,
    including HDFS and BGL system logs.

    HDFS logs primarily represent distributed
    filesystem operational behavior, while BGL
    logs contain high-volume supercomputer
    event logs with anomaly-oriented patterns.

    Multiple datasets were integrated to evaluate:
    <br/><br/>

    • Cross-dataset robustness<br/>
    • Generalization capability<br/>
    • Semantic feature transferability<br/>
    • SOC-oriented anomaly detection reliability
    """,

    body_style
)

elements.append(dataset_heading)

elements.append(
    Spacer(1, 0.2 * inch)
)

elements.append(dataset_text)

elements.append(
    Spacer(1, 0.3 * inch)
)

# ============================================================
# FEATURE ENGINEERING SECTION
# ============================================================

feature_heading = Paragraph(
    "2. Feature Engineering",
    heading_style
)

feature_text = Paragraph(

    """
    Advanced feature engineering was implemented
    to improve anomaly representation quality.

    The feature engineering pipeline included:
    <br/><br/>

    • TF-IDF semantic vectorization<br/>
    • Dataset identity leakage reduction<br/>
    • Structured behavioral feature extraction<br/>
    • IP/path/URL/port detection<br/>
    • Exception and failure keyword extraction<br/>
    • Statistical text behavior analysis<br/>
    • Semantic normalization
    """,

    body_style
)

elements.append(feature_heading)

elements.append(
    Spacer(1, 0.2 * inch)
)

elements.append(feature_text)

elements.append(
    Spacer(1, 0.3 * inch)
)

# ============================================================
# MODEL TRAINING SECTION
# ============================================================

training_heading = Paragraph(
    "3. Model Training and Optimization",
    heading_style
)

training_text = Paragraph(

    """
    Multiple machine learning models were trained
    and evaluated for anomaly detection.

    The evaluation included:
    <br/><br/>

    • Random Forest Classifier<br/>
    • Logistic Regression<br/>
    • Linear SVM<br/>
    • Threshold optimization<br/>
    • Calibration evaluation<br/>
    • Cross-dataset testing<br/>
    • Probability-based scoring
    """,

    body_style
)

elements.append(training_heading)

elements.append(
    Spacer(1, 0.2 * inch)
)

elements.append(training_text)

elements.append(
    Spacer(1, 0.3 * inch)
)

# ============================================================
# GRAPH INSERTION FUNCTION
# ============================================================

def add_graph(title_text, graph_filename):

    graph_path = (
        f"{GRAPH_DIR}/{graph_filename}"
    )

    if os.path.exists(graph_path):

        heading = Paragraph(
            title_text,
            heading_style
        )

        elements.append(heading)

        elements.append(
            Spacer(1, 0.15 * inch)
        )

        img = Image(
            graph_path,
            width=6.3 * inch,
            height=4.5 * inch
        )

        elements.append(img)

        elements.append(
            Spacer(1, 0.3 * inch)
        )

        print(f"✅ Added: {graph_filename}")

    else:

        print(f"⚠ Missing: {graph_filename}")

# ============================================================
# EVALUATION VISUALIZATIONS
# ============================================================

evaluation_heading = Paragraph(
    "4. Evaluation and Visualization",
    heading_style
)

elements.append(evaluation_heading)

elements.append(
    Spacer(1, 0.3 * inch)
)

graphs = [

    ("Confusion Matrix", "confusion_matrix_rf.png"),

    ("ROC Curve Comparison", "roc_curve_comparison.png"),

    ("Feature Importance", "feature_importance_bargraph.png"),

    ("Calibration Curve", "calibration_curve.png"),

    ("Anomaly Score Distribution", "anomaly_score_distribution.png"),

    ("Confidence Scatter Plot", "anomaly_confidence_scatter.png"),

    ("Cross Dataset Evaluation", "cross_dataset_performance.png"),

    ("PCA Embedding Visualization", "pca_embedding_visualization.png"),

    ("t-SNE Embedding Visualization", "tsne_embedding_visualization.png"),

    ("Model Performance Summary", "model_performance_summary.png")
]

for title_text, filename in graphs:

    add_graph(
        title_text,
        filename
    )

# ============================================================
# OBSERVATIONS
# ============================================================

obs_heading = Paragraph(
    "5. Key Observations",
    heading_style
)

obs_text = Paragraph(

    """
    The anomaly detection system demonstrated
    strong same-dataset performance while also
    exposing realistic cross-dataset challenges.

    Major findings included:
    <br/><br/>

    • Strong semantic anomaly discrimination<br/>
    • Effective threshold-based probability scoring<br/>
    • Structured feature contribution improvement<br/>
    • Improved explainability through feature importance<br/>
    • Cross-dataset generalization limitations<br/>
    • Realistic anomaly embedding separation<br/>
    • SOC-oriented severity mapping potential
    """,

    body_style
)

elements.append(obs_heading)

elements.append(
    Spacer(1, 0.2 * inch)
)

elements.append(obs_text)

elements.append(
    Spacer(1, 0.3 * inch)
)

# ============================================================
# CONCLUSION
# ============================================================

conclusion_heading = Paragraph(
    "6. Conclusion",
    heading_style
)

conclusion_text = Paragraph(

    """
    The Adaptive ML-Based Log Anomaly Detection
    System successfully demonstrated the viability
    of combining semantic log analysis, structured
    behavioral extraction, and machine learning
    evaluation for SOC-oriented anomaly detection.

    The final implementation represents a strong
    foundation for future cybersecurity analytics
    platforms and adaptive SOC environments.
    """,

    body_style
)

elements.append(conclusion_heading)

elements.append(
    Spacer(1, 0.2 * inch)
)

elements.append(conclusion_text)

# ============================================================
# BUILD PDF
# ============================================================

print("\nGenerating PDF Report...")

doc.build(elements)

print("\n✅ PDF REPORT GENERATED SUCCESSFULLY")

print("\nSaved At:")
print(PDF_PATH)
```
