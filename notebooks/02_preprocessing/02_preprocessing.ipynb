```python
import pandas as pd

# ==============================
# 1. LOAD PROCESSED DATA
# ==============================
input_path = "E:/LogAnomalyDetector/data/hdfs_processed.csv"
df = pd.read_csv(input_path)

print("Loaded dataset:")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# ==============================
# 2. VALIDATE REQUIRED COLUMNS
# ==============================
required_cols = ["timestamp", "message", "true_label"]

for col in required_cols:
    if col not in df.columns:
        raise ValueError(f"❌ Missing required column: {col}")

# ==============================
# 3. HANDLE MISSING VALUES
# ==============================
print("\nMissing values per column:")
print(df.isna().sum())

before = len(df)
df = df.dropna(subset=["message", "true_label"])
after = len(df)

print(f"\nDropped {before - after} rows due to missing values")
print("Shape after drop:", df.shape)

# ==============================
# 4. CLEAN TEXT (SAFE + CONTROLLED)
# ==============================
df["message"] = df["message"].astype(str)

# Normalize text
df["message"] = df["message"].str.lower()

# Remove extra spaces
df["message"] = df["message"].str.replace(r"\s+", " ", regex=True)

# Trim edges
df["message"] = df["message"].str.strip()

print("\nSample cleaned messages:")
print(df["message"].head())

# ==============================
# 5. FINAL DATASET STRUCTURE
# ==============================
df_final = df[["timestamp", "message", "true_label"]]

# ==============================
# 6. SAVE PREPROCESSED DATA
# ==============================
save_path = "E:/LogAnomalyDetector/data/preprocessed_hdfs.csv"
df_final.to_csv(save_path, index=False)

print("\nSaved preprocessed data to:", save_path)
print("Final shape:", df_final.shape)

print("\nPreview:")
print(df_final.head())
```
