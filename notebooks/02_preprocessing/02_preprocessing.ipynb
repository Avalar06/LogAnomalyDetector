```python
import pandas as pd

input_path = "E:/LogAnomalyDetector/data/labeled_logs.csv"

df = pd.read_csv(input_path)

print("Loaded labeled dataset:")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head())
```

```python
# Check for missing values
print("\nMissing values per column:")
print(df.isna().sum())

# Drop rows with missing message or label
before = len(df)
df = df.dropna(subset=["message", "true_label"])
after = len(df)

print(f"\nDropped {before - after} rows due to missing message/label.")
print("Shape after dropna:", df.shape)
```

```python
# Ensure message is string
df["message"] = df["message"].astype(str)

# Basic normalization
df["message"] = df["message"].str.strip()
df["message"] = df["message"].str.lower()

print("\nSample cleaned messages:")
print(df["message"].head())
```

```python
# Parse timestamp
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

# Drop rows with invalid timestamps
before = len(df)
df = df.dropna(subset=["timestamp"])
after = len(df)

print(f"\nDropped {before - after} rows due to invalid timestamp.")
print("Shape after timestamp parsing:", df.shape)

print("\nTimestamp range:")
print(df["timestamp"].min(), "→", df["timestamp"].max())
```

```python
# Final column order
final_cols = ["timestamp", "message", "true_label"]
df_final = df[final_cols]

save_path = "E:/LogAnomalyDetector/data/preprocessed_logs.csv"
df_final.to_csv(save_path, index=False)

print("\nSaved preprocessed data to:", save_path)
print("Final shape:", df_final.shape)

print("\nFinal dataset preview:")
print(df_final.head())
```
